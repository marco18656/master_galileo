### Ejercicio 4: Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np
# Se definen ambos arrays dados 
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])
# Se calcula la diferencia
diff = a - b
# Calcular la media aritmética de la diferencia
mean_diff = np.mean(diff)
print(f"La media aritmetica de la diferencia es: {mean_diff}")
```

### Ejercicio 1: ¿Cuántos Municipios tienen más de 10,0000 habitantes?

### Ejercicio 2: Realice una gráfica de barras sobre la Población de cada ciudad, ordenela de menor a mayor y responda:
- ¿Cuál es la segunda ciudad más poblada?
- ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
import matplotlib.pyplot as plt
df_ordenado = data.sort_values(by="Poblacion")
# Se genera la grafica de barras
plt.bar(df_ordenado["Municipio"], df_ordenado["Poblacion"])
plt.show()
# Se obtiene la segunda ciudad mas pobladas
segunda_ciudad = df_ordenado.iloc[-2]["Municipio"]
print(f"La segunda ciudad más poblada es {segunda_ciudad}.")
# Se obtiene la posición de Granada en el ranking de ciudades por población
posicion_granada = df_ordenado[df_ordenado["Municipio"] == "Granada"].index[0] + 1
print(f"Granada ocupa la posición {posicion_granada} en el ranking de ciudades más pobladas.")
```

### Ejercicio 4: ¿Cuál es el municipio situado más al Norte? Usar el valor de la coordenada "Y" que representa la latitud en grados. Proporcione también la provincia a la que pertenece y su población.

```python
import pandas as pd
# Se ordenan de forma descendente las latitudes
df_sorted = data.sort_values(by=["Y"], ascending=False)
# Mostrar el primer registro del dataframe ordenado (que corresponderá al municipio más al Norte)
municipio_mas_norte = df_sorted.iloc[0]
print(f" El municipio mas situado al norte es: {municipio_mas_norte['Municipio']}, siendo la provincia de {municipio_mas_norte['Provincia']}, con una población de: {municipio_mas_norte['Poblacion']} habitantes.")
```

### Ejercicio 5: Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada. 

```python
granada_df = data[data["Provincia"] == "Granada"]
# Obtener las estadísticas descriptivas de la población de los municipios de Granada
media = granada_df["Poblacion"].mean()
mediana = granada_df["Poblacion"].median()
desviacion_estandar = granada_df["Poblacion"].std()
valor_maximo = granada_df["Poblacion"].max()
valor_minimo = granada_df["Poblacion"].min()

# Imprimir los resultados
print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Desviación estándar: {desviacion_estandar}")
print(f"Valor máximo: {valor_maximo}")
print(f"Valor mínimo: {valor_minimo}")
```

### Ejercicio 6: Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
# Agrupar los datos por provincia
provincias_df = data.groupby("Provincia")

# Crear un histograma para cada grupo de provincias
for provincia, datos_provincia in provincias_df:
    plt.hist(datos_provincia["Poblacion"], bins=20)
    plt.title(f"Histograma de población en {provincia}")
    plt.xlabel("Población")
    plt.ylabel("Frecuencia")
    plt.show()
```

### Ejercicio 7: Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
# Se seleccionan al azar 50 municipios
municipios_aleatorios = random.sample(set(data["Municipio"]), 50)
# Unicamente se selecciona la data de los 50 municipios
datos_municipios_aleatorios = data[data["Municipio"].isin(municipios_aleatorios)]
# Se calcula el promedio, la desviación estandar de los 50 municipios seleccionados al azar
promedio_poblacion = np.mean(datos_municipios_aleatorios["Poblacion"])
desviacion_estandar_poblacion = np.std(datos_municipios_aleatorios["Poblacion"])
print(f"Promedio de población de los 50 municipios seleccionados: {promedio_poblacion:.2f}")
print(f"Desviación estándar de la población de los 50 municipios seleccionados: {desviacion_estandar_poblacion:.2f}")
```

### Ejercicio 8: Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
import math
def distancia_entre_puntos(x1, y1, x2, y2):
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
import numpy as np
madrid_x, madrid_y = -3.874115128, 40.4720996151
# Calculamos la distancia entre cada municipio y Madrid
distances = np.sqrt((data['X'] - madrid_x)**2 + (data['Y'] - madrid_y)**2)
# Obtenemos el índice del municipio más cercano y más lejano
closest_idx = distances[distances != 0].argmin()
farthest_idx = distances.argmax()
# Obtenemos los nombres de los municipios más cercano y más lejano
closest_municipio = data.loc[closest_idx, 'Municipio']
farthest_municipio = data.loc[farthest_idx, 'Municipio']
# Imprimimos los resultados
print(f"El municipio más cercano a Madrid es: {closest_municipio}")
print(f"El municipio más lejano a Madrid es: {farthest_municipio}")
```
